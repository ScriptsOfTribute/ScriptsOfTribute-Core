```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1 = open('allBotsStatistics.txt', 'r')
Lines = file1.readlines()

data = []
  
data.append(Lines[0].replace('\n', '').replace('SimpleBots.', '').replace('SimpleBotsTests.', '')[:-1].split(','))
for line in Lines[1:]:
    if 'SimpleBots' in line:
        continue
    data.append(line.replace('\n', '').split(','))
```

```python
data = np.array(data)
```

```python
titles = data[0]
```

```python
data = data[1:]
```

```python
data = data.astype(np.float)
```

```python
plt.rcParams["figure.autolayout"] = True
plt.rcParams['figure.figsize'] = 10, 5
x = np.arange(data.shape[0])
for i in range(len(titles)):
    plt.plot(x, np.cumsum(data[:, i],axis=0)/(x +1), label=titles[i])
#plt.plot(x, np.full(len(x), 0.5), color='black')
plt.axhline(y = 0.5, color = 'black')
plt.grid()
plt.legend(loc='lower left', bbox_to_anchor=(1, 0.0))
plt.savefig('simple_bot_clash.png', bbox_inches = "tight")
plt.show()
```
