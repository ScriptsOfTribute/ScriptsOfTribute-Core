```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1 = open('allBotsStatistics.txt', 'r')
Lines = file1.readlines()

data = []
  
data.append(Lines[0].replace('\n', '').replace('SimpleBots.', '').replace('SimpleBotsTests.', '')[:-1].split(','))
for line in Lines[1:]:
    if 'SimpleBots' in line:
        continue
    data.append(line.replace('\n', '').split(','))
```

```python
data
```

```python
data = np.array(data)
```

```python
titles = data[0]
```

```python
data = data[1:]
```

```python
data
```

```python
data = data.astype(np.float)
```

```python
x = np.arange(data.shape[0])
for i in range(len(titles)):
    plt.plot(x, data[:, i], label=titles[i])
plt.legend(loc="best")
plt.show()
```
